14. Agregar marcadores al mapa con las ubicaciones del Colegio Público Manuel Altolaguirre, IES Antonio Dominguez Ortiz y Factoría Cultural - ICAS.

15. Crear un objeto mapa centrado en la ubicación 19.4326, -99.1332 e incluir marcadores para cada una de las ubicaciones del archivo ubicaciones.csv. Cargar el archivo y leer las localizaciones en python, no escribirlas manualmente.

```python
m = folium.Map(location=[19.4326, -99.1332],zoom_start=5)
tooltip = "Click me!"
df = pd.read_csv('ubicaciones.csv')
for i in range(len(df["Latitud"])):
    for j in range(len(df["Longitud"])):
        if i == j:
            folium.Marker([df["Latitud"][i], df["Longitud"][j]], tooltip=tooltip).add_to(m)
m
```
