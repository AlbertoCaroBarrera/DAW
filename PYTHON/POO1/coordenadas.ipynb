Una clase Punto con los siguientes atributos público
- coordX, coordY, coordZ: guardas las coordenadas del punto, como números reales.  
 
Y métodos públicos
- inicializar: establece las coordenadas del punto.
- representar: devuelve una cadena representando el punto.
- sumar: suma al punto el punto del plano especificado.
- multiplicar_por_escalar:multiplica el punto por el número real indicado.
- distancia: devuelve la distancia del punto al punto del plano especificado.

Programa Punto

Inicio Programa
    Clase Punto
    atributos
        publico real: CoordX
        publico real: CoordY
        publico real: CoordZ

    Inicio clase

    público procedimiento inicializar(Punto: propio, Real:x, Real:y, Real:z)
    Inicio procedimiento
        propio.CoordX = x
        propio.CoordeY = y
        propio.CoordZ = z
    Fin procedimiento
    
    público cadena funcion representar(Punto:propio)
    Inicio funcion

    fin funcion

    público procedimiento sumar (Punto:Propio,P)
    Inicio procedimiento
        propio.coordX = propio.coordX + P.coordX
        propio.coordY = propio.coordY + P.coordY
        propio.coordZ = propio.coordZ + P.coordZ
    Fin procedimiento

    publico procedimiento restar (Punto:Propio,P)
    Inicio procedimiento
        propio.coordX = propio.coordX - P.coordX
        propio.coordY = propio.coordY - P.coordY
        propio.coordZ = propio.coordZ - P.coordZ
    Fin procedimiento

    público procedimiento multiplicar(Punto:propio,Real:r)
    inicio procedimiento
        propio.coordX = propio.coordX * r
        propio.coordY = propio.coordY * r
        propio.coordZ = propio.coordZ * r
    fin procedimiento

    público real funcion distancia(Punto:propio,P)
    inicio funcion
        devolver $/sqrt{(propio.coordX - P.coordX)² + {propio.coordY - P.coordY}² + {propio.coordZ - P.coordZ}²}
    fin funcion

```python

import math


class Punto:
    def __init__(self, x=0, y=0, z=0):
        self.coordX = x
        self.coordY = y
        self.coordZ = z

    def __str__(self):
        return str((self.coordX, self.coordY, self.coordZ))

    def sumar(self, P):
        self.coordX += P.coordX
        self.coordY += P.coordY
        self.coordZ += P.coordZ

    def multiplicar(self, r):
        self.coordX *= r
        self.coordY *= r
        self.coordZ *= r

    def distancia(self, P):
        return math.sqrt(
            (P.coordX - self.coordX)**2 +
            (P.coordY - self.coordY)**2 +
            (P.coordZ - self.coordZ)**2
        )
```

```python
import numpy as np


class Vector:
    def __init__(self, A, B):
        self.PuntoA = A
        self.PuntoB = B
        self.coordenadaX = self.PuntoA.coordX-self.PuntoB.coordX
        self.coordenadaY = self.PuntoA.coordY-self.PuntoB.coordY
        self.coordenadaZ = self.PuntoA.coordZ-self.PuntoB.coordZ

    def __str__(self):
        if self.PuntoA == self.PuntoB:
            return 'Ambos puntos son iguales, no generan un vector'
        else:
            return "("+str(self.coordenadaX)+"," + str(self.coordenadaY)+"," + str(self.coordenadaZ)+")"

    def producto_escalar(self, V):
        return self.coordenadaX*V.coordenadaX+self.coordenadaY*V.coordenadaY+self.coordenadaZ*V.coordenadaZ

    def modulo(self):
        return self.PuntoA.distancia(self.PuntoB)

    def producto_vectorial(self, V):
        A = [self.coordenadaX, self.coordenadaY, self.coordenadaZ]
        B = [V.coordenadaX, V.coordenadaY, V.coordenadaZ]
        a, b, c = np.cross(A, B)
        return Vector(Punto(), Punto(a, b, c))

    def ortogonal(self, V):
        if self.producto_escalar(V) == 0:
            return 'Los vectores son ortogonales'
        else:
            return 'Los vectores no son ortogonales'
```
