Una clase Matriz con los siguientes atributos privados:
- elementos: guarda los elementos de la matriz, como una lista de filas, y cada fila como una lista de números reales.   
y métodos públicos:
- inicializar: dados el número de filas y columnas de la matriz, pone a cero todos sus elementos.
- representar:devuelve una cadena que representa la matriz.
- obtener_numero_filas, obtener_numero_columnas: devuelve el número de filas  y columnas, respectivamente, de la matriz.
- obtener_elemento: devuelve el elemento de la matriz que ocupa la posición especificada.
- cambiar_elemento: cambia el elemento que se encuentra en la posición especificada por el nuevo elemento indicado.
- redimensionar: cambia las dimensiones de la matriz por las especificadas, poniendo además todos sus elementos a cero.    
Una Clase MatrizConOperaciones que herede de la clase anterior y añada los siguientes métodos públicos:
- sumar: añade a la matriz, si se puede, la matriz especificada.
- multiplicar_por_escalar: multiplica la matriz por el número real especificado.
- multiplicar: multiplica la matriz, si se puede, por la matriz especificada.
- transponer:transponer la matriz.  
Una clase MatrizCuadrada que hereda de la anterior y reescribe el siguiente método público
- inicializar: dada la dimensión de la matriz, pone a cero todos sus elementos.
- redimensionar: cambia las dimensiones de la matriz por la especificada, poniendo además todos sus elementos a cero.  
y añade los siguientes métodos públicos:
- obtener_matriz_adjunta: devuelve la matriz adjunta de la posición especificada.
- calcular_determinante: devuelve el determinate de la matriz.

```python

def fila(l):
    cadena = ''
    while len(l) > 0:
        cadena += str(l[0]) + ' '
        l = l[1:]
    return cadena
def rep(l):
    cadena = ''
    for i in range(len(l)):
        cadena += (fila(l[i])) + '\n'
    return cadena
```

```python
def matriz_zero(fila, columna):
    matriz = []
    for i in range(fila):
        matriz.append([])
        for j in range(columna):
            matriz[i].append(0)
    return matriz
```

```python
class Matriz:
    def __init__(self, nf,nc):
        self.__elementos = matriz_zero(nf,nc)

    def __str__(self):
        return rep(self.__elementos)
    
    def obtener_numero_filas(self):
        return len(self.__elementos)
    
    def obtener_numero_columnas(self):
        return len(self.__elementos[0])
    
    def obtener_elemento(self, fila, columna):
        return self.__elementos[fila][columna]
    
    def cambiar_elemento(self, fila, columna, valor):
        self.__elementos[fila][columna] = valor
        
    def redimensionar(self, nfilas, ncolumnas):
        self.__elementos = matriz_zero(nfilas,ncolumnas)
    
class MatrizConOperaciones(Matriz):
    def __init__(self, nf,nc):
        self.__elementos = matriz_zero(nf,nc)
    
    
    def numerofilas(self):
        return len(self.__elementos)
    
    
    def sumar(self,matriz2):
        for i in range(self.obtener_numero_filas()):
            for j in range(self.obtener_numero_columnas()):
                self.cambiar_elemento(i,j,self.obtener_elemento(i,j)+matriz2.obtener_elemento(i,j))
                
    def multiplicar_por_escalar(self,escalar):
        for i in range(self.obtener_numero_filas()):
            for j in range(self.obtener_numero_columnas()):
                self.cambiar_elemento(i,j,self.obtener_elemento(i,j)*escalar)
    
    def multiplicar(self,matriz2):
        nf1 = self.obtener_numero_filas()
        nc1 = self.obtener_numero_columnas()
        nf2 = matriz2.obtener_numero_filas()
        nc2 = matriz2.obtener_numero_columnas()
        
        for i in range(nf1):
            for j in range(nc2):
                acumulador = 0
                for k in range(nc1):
                    acumulador += self.obtener_elemento(i, k) * matriz2.obtener_elemento(k, j)
                self.cambiar_elemento(i,j,acumulador)
            
    def transponer(self):
        traspuesta = MatrizConOperaciones(self.obtener_numero_columnas(),self.obtener_numero_filas())
        
        for i in range(self.obtener_numero_filas()):
            for j in range(self.obtener_numero_columnas()):
                traspuesta.cambiar_elemento(i,j,self.obtener_elemento(j,i))
        
        for i in range(traspuesta.obtener_numero_filas()):
            for j in range(traspuesta.obtener_numero_columnas()):
                self.cambiar_elemento(i,j,traspuesta.obtener_elemento(i,j))
                
```

```python
M = Matriz(2,3)
print(M)
print(M.obtener_numero_filas())
print(M.obtener_numero_columnas())
M.cambiar_elemento(1,1,2)
print(M)


M2= MatrizConOperaciones(3,3)
M2.cambiar_elemento(1,1,2)
M2.cambiar_elemento(0,1,2)
print(M2)
M3=MatrizConOperaciones(3,3)
M3.cambiar_elemento(1,1,2)
M2.sumar(M3)
print(M2)
M2.multiplicar_por_escalar(2)
print(M2)
M2.multiplicar(M3)
print(M3)
print(M2)
```

```python
#Sin terminar, la idea es para una matriz de dim 1 y 2 calculamos adjuntos(cofactores) y su determinante. Para el resto calculamos mediante recursion.

class MatrizCuadrada(MatrizConOperaciones):

    def matriz_adjunta(self,i,j):
        if self.obtener_numero_filas() == 3:
            return (-1)^(i+j)*self.determinante()
    
    def determinatente(self):
        if self.obtener_numero_filas() == 2:
            return self.obtener_elemento(0,0)*self.obtener_elemento(1,1)+self.obtener_elemento(0,1)*self.obtener_elemento(1,0)
        else:
            pass
        
M4 = MatrizCuadrada(2,2)
M4.cambiar_elemento(0,0,1)
M4.cambiar_elemento(1,1,2)
print(M4)
print(M4.determinatente())
print(M4.matriz_adjunta(0,0))
```

```python
#Determinante para una lista de lista

def determinante(matriz):
    n = len(matriz)
    if n == 1:
        return matriz[0][0]
    elif n == 2:
        return matriz[0][0] * matriz[1][1] - matriz[0][1] * matriz[1][0]
    else:
        det = 0
        for j in range(n):
            signo = (-1) ** j
            submatriz = []
            for i in range(1, n):
                fila = []
                for k in range(n):
                    if k != j:
                        fila.append(matriz[i][k])
                submatriz.append(fila)
            subdet = determinante(submatriz)
            det += signo * matriz[0][j] * subdet
        return det
```
