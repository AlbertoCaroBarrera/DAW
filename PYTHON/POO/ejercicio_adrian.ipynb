```python
class Producto:
    def __init__(self, name, stock,price):
        self.nombre = name
        self.stock = stock
        self.precio = price

    def __str__(self):
        return f'Nombre: {self.nombre} \nStock: {self.stock} \nPrecio: {self.precio}€'
    
    def cambiar_stock(self, cantidad):
        self.stock += cantidad

    def cambiar_precio(self, precio):
        self.precio = precio
        
leche = Producto('leche', 10, 1.25)
print(leche)
```

    publico procedimiento añadir_producto(Tienda:Propio, producto:Producto,cantidad:int)

    Inicio procedimiento

    Fin procedimiento



```python
class Tienda:
    def __init__(self,name,dir,telf,l1=[],l2=[],dic={}):
        self.nombre = name
        self.direccion = dir
        self.telefono = telf
        self.__clientes = l1
        self.__empleados = l2
        self.productos = dic

        for producto,stock in self.productos.items():
            producto.stock = stock


    def __str__(self):
        
        cadena_clientes = ""
        for cliente in self.__clientes:
            cadena_clientes += str(cliente)
            
        cadena_empleados = ""
        for empleado in self.__empleados:
            cadena_empleados += str(empleado)
        
        cadena_productos = ""
        for producto in self.productos:
            cadena_productos += str(producto)

        return f'Nombre: {self.nombre}\nDireccion: {self.direccion}\nTelefono: {self.telefono}\n\nClientes:\n'+cadena_clientes+'\nEmpleados:\n'+cadena_empleados+'\nProductos:\n'+cadena_productos
        
```
